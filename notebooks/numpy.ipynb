```python
import numpy as np
a = np.array([
    [1,2,3],
    [4,5,6]
])
b = np.array([
    [7,8,9],
    [10,11,12]
])

c = np.add(a,b)
d = np.subtract(a,b)
e = a*b #multiply
f = a/b #divide
g = a%b #mod(a,b)
h = np.pow(a,b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
```

```python
import numpy as np
a = np.array([
    [9,4,3],
    [1,7,6],
    [8,2,5]
    ])
print(a)
print("minimum: ", np.amin(a))
print("column_min: ", np.amin(a,axis=0))
print("row_min: ", np.amin(a,axis=1))
print("maximum: ", np.amax(a))
print("column_max: ", np.amax(a,axis=0))
print("row_max: ", np.amax(a,axis=1))
```

```python
import numpy as np
a = np.array([
    [9,4,3],
    [1,7,6],
    [8,2,5]
])
print(a)
print("mean: ",np.mean(a))
print("col_mean: ",np.mean(a,axis=0))
print("row_mean: ",np.mean(a,axis=1))
```

```python
import numpy as np
a = np.array([
    [9,4,3],
    [1,7,6],
    [8,2,5]
])
print(a)
print("median: ",np.median(a))
print("col_median: ",np.median(a,axis=0))
print("row_median: ",np.median(a,axis=1))
```

```python
import numpy as np
a = np.array([
    [9,4,3],
    [1,7,6],
    [8,2,5]
])
print(a)
print("variance = ",np.var(a))
print("Std_deviation = ",np.std(a))
```

```python
import numpy as np
a = np.array([4,6,5,2,1,9,0])
print("1D Unsorted = \n",a)
print("1D Sorted = \n",np.sort(a))
b = np.array([
    [9,4,3],
    [1,7,6],
    [8,2,5]
])
print("2D Unsorted = \n",b)
print("2D Sorted = \n",np.sort(b))
```

```python
import numpy as np
a = np.array([
    [1,2,3],
    [4,5,6]
])
b = np.array([
    [10,20,30],
    [40,50,60]
])
c = np.concatenate((a,b)) #default
d = np.concatenate((a,b),axis=1)
print("col_concat: \n",c)
print("row_concat: \n",d)
```
